```python
class Solution(object):
    def validateStackSequences(self, pushed, popped):
        """
        :type pushed: List[int]
        :type popped: List[int]
        :rtype: bool
        """
        stack = []
        que_pushed = 0
        que_popped = 0
        que_stack = -1

        isSquence = False

        while  que_pushed <= len(pushed) and que_popped <= len(popped) :

            print(que_pushed == (len(pushed)-1))
            print(que_popped == (len(popped)-1))

            if(que_pushed == (len(pushed)) and que_popped == (len(popped)) and que_stack == -1):
                isSquence = True
                break

            has_pushed = False
            has_popped = False

            if(que_pushed < len(pushed)):
                print("Que pushed:" , que_pushed)
                stack.append(pushed[que_pushed])
                que_pushed += 1
                has_pushed = True
                
            if (que_stack != -1):
                if (stack[que_stack] == popped[que_popped]):
                    print("Que Stack in pop:" , que_stack)
                    print("Que popped:" , que_popped)
                    stack.pop(que_stack)
                    que_popped += 1
                    has_popped = True

           

            if(has_pushed and not has_popped):
                # print("Que Stack:" , que_stack)
                que_stack += 1

            if(not has_pushed and has_popped):
                # print("Que Stack:" , que_stack)
                que_stack -= 1
            
            print("Que pushed", que_pushed)
            print("Que Stack:" , que_stack)    
            print("Current Stack",stack)

        
        return isSquence
```
