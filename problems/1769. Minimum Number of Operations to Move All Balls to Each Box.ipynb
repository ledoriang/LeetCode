Class definition

```python
"""
Runtime Complexity - O(n^2)
"""
class Solution:
    def minOperations(self, boxes: str) -> list[int]:
        totalCount = []
        for i in range(len(boxes)):
            moveCount = 0
            for ii in range(len(boxes)):
                if(i != ii):
                    if(int(boxes[ii]) != 0):
                        moveCount += abs(ii - i)
            totalCount.append(moveCount)
        return totalCount
        
```

```python
sol = Solution()
sol.minOperations(boxes=boxes)
```
